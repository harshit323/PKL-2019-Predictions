**TASK 5 <br>
Predict the team with the highest super-performance total.** <br>
S.P.T. = Total number of super-raids in the tournament + total number of super-tackles in the tournament + total number of all-outs inflicted in the tournament - total number of all-outs conceded in the tournament

```python

def get_dataframe(sp,s,points_col_name):
    team_name = [] 
    matches_played = [] 
    points = [] 
    
    l = sp.findAll('div',attrs={'class':'lb-detailItem-wrap'})
    
    for x in l:
        n = x.find('div', attrs={'class':'sipk-lb-playerName'})
        m = x.find('div', attrs={'class':'sipk-lb-matchedPlayed'})
        p = x.find('div', attrs={'class':'sipk-lb-raidPoints'})
        team_name.append(n.text.strip())
        matches_played.append(m.text.strip())
        points.append(p.text.strip())

    frame = pd.DataFrame({'Team':team_name,s+'Matches_Played':matches_played,s+points_col_name:points})
    return frame

```
